```python
os.listdir(data_path)
```

```python
# load the data
data = pd.read_csv(data_path+'/list_attribute.txt', delim_whitespace=True, skiprows=1)
data.head()
```

```python
data = data.reset_index()
data = data.rename(columns={'index': 'image'})
data.head()
```

```python
data = data[['image','Male']]
data
```

```python
data['Male'] = data['Male'].replace(-1, 0)
data
```

```python
images_list = os.listdir(data_path+'/Images')
```

```python
# filter the data with available images only (5000 images)
condition = data['image'].isin(images_list)
data = data[condition]
data = data.reset_index(drop=True)
data
```

```python
data.to_csv('gender.csv', index=False)
```
